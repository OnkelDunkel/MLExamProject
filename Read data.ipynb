```python
df = df[df["work-balance-stars"] != "none"]
df = df[df["culture-values-stars"] != "none"]
df = df[df["carrer-opportunities-stars"] != "none"]
df = df[df["comp-benefit-stars"] != "none"]
df = df[df["senior-mangemnet-stars"] != "none"]
df.count()[0]
```

```python
df = df[["job-title", "work-balance-stars","culture-values-stars","carrer-opportunities-stars","comp-benefit-stars","senior-mangemnet-stars","company"]]
df.head()
```

```python
df["company"].value_counts()
```

```python
df_job_split = pd.DataFrame(df["job-title"].str.split(" - ", 1).tolist(), columns = ["current","job"])
df["job-title"] = df_job_split.job
df["current"] = df_job_split.current
```
